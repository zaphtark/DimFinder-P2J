## Load JSON File

```python
import json
```

```python
with open("Lysistrata.json", "r", encoding="utf8") as file:
    whole_play = json.load(file)
```

## Get all lines

```python
all_lines = [] #this will contain both the speaker and the text content

for said in whole_play["content"]:
    line = {'name':said['name'], 'text':[]}
    for verse in said['text']:
        line['text'].append(verse['text'])
    all_lines.append(line)
```

```python
all_lines_text = []  #this is a list of only the text content

for line in all_lines:
    for verse in line['text']:
        all_lines_text.append(verse)
```

## Get lines by character

```python
text_by_char = {}

for line in all_lines:
    for verse in line['text']:
        if line['name'] in text_by_char.keys():
            text_by_char[line['name']].append(verse)
        else:
            text_by_char[line['name']] = [verse]
```

## Clean text

```python
from cltk import NLP

nlp = NLP("grc")
```

```python
from unicodedata import normalize

def clean_texts(list_of_texts):
    purge_char = [",", ".", ";", ":", "—", "†"] #,"'" Enlever apostrophe si on lemattise
    clean_texts = list()
    for text in list_of_texts:
        clean_text = text.replace("\n", " ")
        for char in purge_char:
            clean_text = clean_text.replace(char, "")
        cltk_doc = nlp.analyze(clean_text)
        clean_text = " ".join([word.lemma for word in cltk_doc.words]) #change to word.lemma for lemmatization
        #print([word.lemma for word in cltk_doc.words])
        #input()
        #clean_text = normalize("NFKD", clean_text).translate({ord(c): None for c in "̓̔́̀͂̈ͅ"})
        clean_texts.append(clean_text)
    return clean_texts
```

## Compare similarity

```python
from scipy.spatial.distance import euclidean, cosine
```

```python
def calculate_average_similarity(line_vector, all_vectors):
    distance_sum = 0
    for idx, vector in enumerate(all_vectors):
        vector_array = vector.toarray()[0]
        vector_line_array = line_vector.toarray()[0]
        sim = euclidean(vector_array, vector_line_array)
        distance_sum += float(sim)
    avg = distance_sum / all_vectors.shape[0]
    return avg
```

```python
def calculate_character_similarity(char_vectors, compare_vectors):
    distance_sum = 0
    for vector in char_vectors:
        distance_sum += calculate_average_similarity(vector, compare_vectors)
    avg = distance_sum / char_vectors.shape[0]
    return avg
```

## Guess who

```python
def guess_line(line):
    current_result = {'name':'None','similarity':0.0}
    for key in all_characters.keys():
        line_vectors = full_play_vectorizer.transform(line)
        line_char_similarity = calculate_average_similarity(line_vectors,all_characters[key])
        if line_char_similarity > current_result['similarity']:
            current_result['name'] = key
            current_result['similarity'] = line_char_similarity
    return current_result
```

```python
def guess_who(char):
    current_result = {'name':'None','similarity':0.0}
    for key in all_characters.keys():
        char_vectors = full_play_vectorizer.transform(char)
        char_similarity = calculate_character_similarity(char_vectors,all_characters[key])
        if char_similarity > current_result['similarity']:
            current_result['name'] = key
            current_result['similarity'] = char_similarity
    return current_result
```
